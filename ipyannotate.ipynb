**importing necassery packages**

```python
from ipyannotate import annotate
from google.colab import output
output.enable_custom_widget_manager()
from glob import glob
from PIL import Image
from ipyannotate.buttons import ValueButton as Button
```

```python
from ipyannotate.buttons import (
    ValueButton as Button,
    NextButton as Next,
    BackButton as Back
)

data = [Image.open(_) for _ in glob('images/*.jpg')]
annotation = annotate(data, buttons=[Button('bat'), Button('eagle'),Button('tiger'), Back(), Next()])
annotation
```

```python
annotation.tasks
```

```python
labels = [task.value for task in annotation.tasks] # get labels
labels
```
